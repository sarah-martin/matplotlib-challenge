```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/mouse_metadata.csv"
study_results_path = "data/study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_complete_df = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
data_complete_df
```

```python
# Checking the number of mice in the DataFrame.
len(data_complete_df["Mouse ID"].value_counts())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate=data_complete_df[data_complete_df.duplicated()]
duplicate
```

```python
# Get all the data for the duplicate mouse ID. 
duplicate_data=data_complete_df.loc[(data_complete_df["Mouse ID"]=="g989")]
duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_data_df= data_complete_df[data_complete_df['Mouse ID'] != "g989"]
cleaned_data_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_data_df["Mouse ID"].value_counts())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Group by regimen
grouped_stats=cleaned_data_df.groupby(["Drug Regimen"])

# Use calculation functions on tumor volume for the grouped data
tumor_mean=grouped_stats["Tumor Volume (mm3)"].mean()
tumor_median=grouped_stats["Tumor Volume (mm3)"].median()
tumor_variance=grouped_stats["Tumor Volume (mm3)"].var()
tumor_stdv=grouped_stats["Tumor Volume (mm3)"].std()
tumor_sem=grouped_stats["Tumor Volume (mm3)"].sem()

# Create the dataframe
regimen_summary_df=pd.DataFrame({"Mean Tumor Volume":tumor_mean,
                                 "Median Tumor Volume":tumor_median,
                                 "Tumor Volume Variance":tumor_variance,
                                 "Tumor Volume Std. Dev.":tumor_stdv,
                                 "Tumor Volume Std. Err.":tumor_sem})

regimen_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Create dataframe only consisting of final timepoints for each mouse by dropping all duplicates
final_tumor_df=cleaned_data_df.drop_duplicates(subset="Mouse ID",keep="last")
final_tumor_df
```

```python
# Quantitatively determine capomulin outliers

#Create campomulin dataframe
cap_df = final_tumor_df[(final_tumor_df["Drug Regimen"] == "Capomulin")]

# Find outliers
cap_quartiles=cap_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
cap_lowerq=cap_quartiles[0.25]
cap_upperq=cap_quartiles[0.75]
cap_iqr=cap_upperq-cap_lowerq
cap_lower_bound=cap_lowerq-(1.5*cap_iqr)
cap_upper_bound=cap_upperq+(1.5*cap_iqr)
```

```python
# Quantitatively determine ramicane outliers

#Create ramicane dataframe
ram_df = final_tumor_df[(final_tumor_df["Drug Regimen"] == "Ramicane")]

# Find outliers
ram_quartiles=ram_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
ram_lowerq=ram_quartiles[0.25]
ram_upperq=ram_quartiles[0.75]
ram_iqr=ram_upperq-ram_lowerq
ram_lower_bound=ram_lowerq-(1.5*ram_iqr)
ram_upper_bound=ram_upperq+(1.5*ram_iqr)
```

```python
# Quantitatively determine infubinol outliers

#Create infubinol dataframe
inf_df = final_tumor_df[(final_tumor_df["Drug Regimen"] == "Infubinol")]

# Find outliers
inf_quartiles=inf_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
inf_lowerq=inf_quartiles[0.25]
inf_upperq=inf_quartiles[0.75]
inf_iqr=inf_upperq-inf_lowerq
inf_lower_bound=inf_lowerq-(1.5*inf_iqr)
inf_upper_bound=inf_upperq+(1.5*inf_iqr)
```

```python
# Quantitatively determine ceftamin outliers

#Create ceftamin dataframe
cef_df = final_tumor_df[(final_tumor_df["Drug Regimen"] == "Ceftamin")]

# Find outliers
cef_quartiles=cef_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
cef_lowerq=cef_quartiles[0.25]
cef_upperq=cef_quartiles[0.75]
cef_iqr=cef_upperq-cef_lowerq
cef_lower_bound=cef_lowerq-(1.5*cef_iqr)
cef_upper_bound=cef_upperq+(1.5*cef_iqr)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Grab tumor volume in each regimen dataframe
cap_plot=cap_df["Tumor Volume (mm3)"]
ram_plot=ram_df["Tumor Volume (mm3)"]
inf_plot=inf_df["Tumor Volume (mm3)"]
cef_plot=cef_df["Tumor Volume (mm3)"]

# Plot tumor volume per regimen and label appropriately 
vol_plot=[cap_plot,ram_plot,inf_plot,cef_plot]
labels = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
plt.boxplot(vol_plot, labels=labels, flierprops=dict(markersize=12, markerfacecolor='red'))
plt.xlabel("Regimen")
plt.ylabel("Final Tumor Volume")
plt.title("Final Timepoint Tumor Volume Per Regimen")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

# Create dataframe for single mouse treated by Capomulin
mouse_s185=cleaned_data_df.loc[cleaned_data_df["Mouse ID"]=="s185"]

# Establish axis and plot then label appropriately 
s185_x=mouse_s185["Timepoint"]
s185_y=mouse_s185["Tumor Volume (mm3)"]
plt.plot(s185_x,s185_y)
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment of Mouse s185")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

#Create a complete capomulin dataframe then group by Mouse ID and average
cap_weight_df= cleaned_data_df[cleaned_data_df["Drug Regimen"] == "Capomulin"]
cap_gb_avg = cap_weight_df.groupby(["Mouse ID"]).mean()

#Establish axis, plot, and label appropriately 
cap_x = cap_gb_avg["Weight (g)"]
cap_y= cap_gb_avg["Tumor Volume (mm3)"]
plt.scatter(cap_x, cap_y)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Average Tumor Volume")
plt.show()
```
